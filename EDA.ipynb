```python

import pandas as pd
import numpy as np
import seaborn as sns
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
import pickle5 as pickle
four_years_df = pd.read_pickle("cleaned_df.pkl")
```

 # 1. Does BPM differ depending on minutes played? Are outliers in BPM related to minutes played?

- Ans: Outliers tend to exist in players that has played less than 100 minutes

```python
four_years_df['MP_quartiles'] = pd.qcut(four_years_df['MP'], 10)
four_years_df.boxplot(column='BPM',by='MP_quartiles',figsize=(20,15))
```

# 2. What are some of the features that has most correlated with our target BPM or Average BPM?

- Ans: Non of the variables seem to be very correlated. This could mean that relationship between the features and target is not linear.

```python
corr_matrix = four_years_df.corr()
corr_matrix.style.background_gradient(cmap='coolwarm')
```

# 3. Does BPM differ accross positions?

- Ans: BPM tend to remain constant across player positions.

```python
four_years_df.boxplot(column='BPM',by='Pos',figsize=(7,5))

```
